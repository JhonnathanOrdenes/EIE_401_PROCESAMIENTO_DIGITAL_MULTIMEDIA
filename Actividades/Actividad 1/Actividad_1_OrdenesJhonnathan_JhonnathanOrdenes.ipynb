```python
import numpy as np
import matplotlib.pyplot as plt

# Se inician los espacios vectoriales para las variables x, y, z respectivamente desde 0 a 50 con una resolucion de 1000 puntos por intervalo
x = np.linspace (0,50,1000)
y = np.linspace (0,50,1000)
z = np.linspace (0,50,1000)

funcion1 = np.exp(-(x**2))# Se establece la funcion F(x)
result2 = np.sin(x) # Se establece la parte senoidal de la funcion F(z)
funcion2 = np.multiply(y,result2) # Se multiplica la parte senoidal de F(z) por y para obtener la funcion F(z) completa

# Se inicia la etapa de visualizacion de los graficos de ambas funciones
plt.subplot(1,2,1) # El subplot corresponde a el grafico de la funcion F(x)
plt.plot (x,funcion1)
plt.xlim(0,4) # Se establece el limite del grafico a 4 ya que en ese tiempo la señal de la funcion F(x) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(x)")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.subplot(1,2,2) # El subplot corresponde a el grafico de la funcion F(z)
plt.plot (z,funcion2)
plt.xlim(0,30) # Se establece el limite del grafico a 30 ya que en ese tiempo la señal de la funcion F(z) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(z)")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.show
```

### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np 
import matplotlib.pyplot as plt

# Se inician los espacios vectoriales para las variables x, y, z respectivamente desde 0 a 50 con una resolucion de 1000 puntos por intervalo
x = np.linspace (0,50,1000)
y = np.linspace (0,50,1000)
z = np.linspace (0,50,1000)

ruidoG1 =np.random.normal(0,0.1,len(x)) # Se genera un ruido gaussiano con una desviacion estandar de 0.1 a fin de demostrar como afecta esta la señal sin sobresaturar el resultado final de la funcion F(x)
ruidoG =np.random.normal(0,1,len(y)) # Se genera un ruido gaussiano con una desviacion estandar de 1 a fin de demostrar como esta afecta la señal de la funcion F(z)
funcion1 = np.exp(-(x**2)) # Se establece la funcion F(x)
result2 = np.sin(x) # Se establece la parte senoidal de la funcion F(z)
funcion2 = np.multiply(y,result2) # Se multiplica la parte senoidal de F(z) por y para obtener la funcion F(z) completa
funcionR1 = ruidoG1 + funcion1 # Se le suma el ruido gaussiano a F(x)
funcionR2 = ruidoG + funcion2 # Se le suma el ruido gaussiano a F(z)

# Se inicia la etapa de visualizacion de los graficos de ambas funciones
plt.subplot(1,2,1) # el subplot corresponde a el grafico de la funcion F(x)
plt.plot (x,funcionR1)
plt.xlim(0,6) # Se establece el limite del grafico a 6 ya que en ese tiempo la señal de la funcion F(x) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(x) con ruido gaussiano")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.subplot(1,2,2)# el subplot corresponde a el grafico de la funcion F(z)
plt.plot (z,funcionR2)
plt.xlim(0,30) # Se establece el limite del grafico a 30 ya que en ese tiempo la señal de la funcion F(z) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(z) con ruido gaussiano")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.show
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa es un tipo de señal, que se caracteriza por tener una distribución de energía la cual disminuye a medida que se aumenta la frecuencia de esta, teniendo un comportamiento inversamente proporcional entre la intensidad de este con su frecuencia, en consecuencia las bajas frecuencias presentan una alta cantidad de energía, a diferencia del ruido blanco, el cual presenta una distribución de energía pareja a lo largo del espectro de frecuencia

A fin de demostrar la diferencia del comportamiento de ambos tipos de ruido, es que, a continuación, se ilustran ambos tipos de ruido.

```python
import numpy as np
import matplotlib.pyplot as plt
# Funcion que establece a el valor de f definido como una transformada de fourier 
#y entrega valores para graficar el comportamiento del ruido a procesar
def plot_spectrum(s):
    f = np.fft.rfftfreq(len(s)) 
    return plt.loglog(f, np.abs(np.fft.rfft(s)))[0]

# Funcion que entrega como resultado el comportamiento del ruido blanco
def noise_psd(N, psd = lambda f: 1):
        X_white = np.fft.rfft(np.random.randn(N));
        S = psd(np.fft.rfftfreq(N))
        # Normalize S
        S = S / np.sqrt(np.mean(S**2))
        X_shaped = X_white * S;
        return np.fft.irfft(X_shaped);

# Funcion que hace llamado al tipo de color a procesar
def PSDGenerator(f):
    return lambda N: noise_psd(N, f)

@PSDGenerator
def white_noise(f): # Si se necesita obtener el ruido blanco, se retorna a la funcion "noise_psd"
    return 1;

@PSDGenerator
def pink_noise(f): # Si se necesita obtener el ruido rosa, se retorna a la funcion "noise_psd" dividida por la raiz cuadrada de "f"
    return 1/np.where(f == 0, float('inf'), np.sqrt(f))

#Grafico que detalla el comportamiento del ruido rosa
plt.style.use('default')
plt.figure(figsize=(12, 8), tight_layout=True)
for G, c in zip(
        [pink_noise], 
        ['hotpink']):
    plot_spectrum(G(30*50_000)).set(color=c, linewidth=3)
plt.suptitle("Ruido Rosa");
plt.xlabel("Frecuencia")
plt.ylabel("Amplitud")
plt.grid()
plt.ylim([1e-1, None]);

#Grafico que detalla el comportamiento del ruido blanco
plt.style.use('default')
plt.figure(figsize=(12, 8), tight_layout=True)
for G, c in zip(
        [white_noise], 
        ['blue']):
    plot_spectrum(white_noise(30*50_000)).set(color=c, linewidth=3)
plt.suptitle("Ruido Blanco");
plt.xlabel("Frecuencia")
plt.ylabel("Amplitud")
plt.grid()
plt.ylim([1e-1, None]);
```

De las figuras obtenidas, se observa lo previamente descrito, confirmando efectivamente que la amplitud del ruido blanco disminuye a medida que aumenta su frecuencia a diferencia del ruido blanco, el cual se mantiene casi parejo en su amplitud durante todo el espectro de frecuencia.
