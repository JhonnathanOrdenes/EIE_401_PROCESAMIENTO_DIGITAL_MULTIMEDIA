## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2} 
$$

$$
f(z) =  ysin(x) 
$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Se inician los espacios vectoriales para las variables x, y, z respectivamente desde 0 a 50 con una resolucion de 1000 puntos por intervalo
x = np.linspace (0,50,1000)
y = np.linspace (0,50,1000)
z = np.linspace (0,50,1000)

funcion1 = np.exp(-(x**2))# Se establece la funcion F(x)
result2 = np.sin(x) # Se establece la parte senoidal de la funcion F(z)
funcion2 = np.multiply(y,result2) # Se multiplica la parte senoidal de F(z) por y para obtener la funcion F(z) completa

# Se inicia la etapa de visualizacion de los graficos de ambas funciones
plt.subplot(1,2,1) # El subplot corresponde a el grafico de la funcion F(x)
plt.plot (x,funcion1)
plt.xlim(0,4) # Se establece el limite del grafico a 4 ya que en ese tiempo la señal de la funcion F(x) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(x)")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.subplot(1,2,2) # El subplot corresponde a el grafico de la funcion F(z)
plt.plot (z,funcion2)
plt.xlim(0,30) # Se establece el limite del grafico a 30 ya que en ese tiempo la señal de la funcion F(z) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(z)")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.show
```

### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np 
import matplotlib.pyplot as plt
# Se inician los espacios vectoriales para las variables x, y, z respectivamente desde 0 a 50 con una resolucion de 1000 puntos por intervalo
x = np.linspace (0,50,1000)
y = np.linspace (0,50,1000)
z = np.linspace (0,50,1000)

ruidoG1 =np.random.normal(0,0.1,len(x)) # Se genera un ruido gaussiano con una desviacion estandar de 0.1 a fin de demostrar como afecta esta la señal sin sobresaturar el resultado final de la funcion F(x)
ruidoG =np.random.normal(0,1,len(y)) # Se genera un ruido gaussiano con una desviacion estandar de 1 a fin de demostrar como esta afecta la señal de la funcion F(z)
funcion1 = np.exp(-(x**2)) # Se establece la funcion F(x)
result2 = np.sin(x) # Se establece la parte senoidal de la funcion F(z)
funcion2 = np.multiply(y,result2) # Se multiplica la parte senoidal de F(z) por y para obtener la funcion F(z) completa
funcionR1 = ruidoG1 + funcion1 # Se le suma el ruido gaussiano a F(x)
funcionR2 = ruidoG + funcion2 # Se le suma el ruido gaussiano a F(z)

# Se inicia la etapa de visualizacion de los graficos de ambas funciones
plt.subplot(1,2,1) # el subplot corresponde a el grafico de la funcion F(x)
plt.plot (x,funcionR1)
plt.xlim(0,6) # Se establece el limite del grafico a 6 ya que en ese tiempo la señal de la funcion F(x) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(x) con ruido gaussiano")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.subplot(1,2,2)# el subplot corresponde a el grafico de la funcion F(z)
plt.plot (z,funcionR2)
plt.xlim(0,30) # Se establece el limite del grafico a 30 ya que en ese tiempo la señal de la funcion F(z) se establece o no ofrece mayor informacion necesaria pasada ese punto
plt.title("f(z) con ruido gaussiano")
plt.xlabel("tiempo")
plt.ylabel("Amplitud")
plt.grid()
plt.show
```
